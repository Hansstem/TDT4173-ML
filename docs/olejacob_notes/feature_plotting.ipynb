```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt

import sys
sys.path.append('../..')  # Add the parent directory (project root) to the sys.path

from model_notebooks.data_func import read_data

# Now you can import and use the read_data module



dataframes = read_data.get_training_data()
X_frames_train = dataframes[0]
Y_frames_train = dataframes[1]
X_frames_test = read_data.get_test_data()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(20, 10), sharex=True)
feature_name = 'absolute_humidity_2m:gm3'
X_frames_train[0][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test A', color='blue')
X_frames_test[0][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test  A', color='green')
Y_frames_train[0][['time', 'pv_measurement']].set_index('time').plot(ax=axs[1], title='Targets  A', color='red')
```

For dataset A we observe that the data is complete.

```python
fig, axs = plt.subplots(2, 1, figsize=(20, 10), sharex=True)
feature_name = 'absolute_humidity_2m:gm3'
X_frames_train[1][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test  B', color='blue')
X_frames_test[1][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test  B', color='green')
Y_frames_train[1][['time', 'pv_measurement']].set_index('time').plot(ax=axs[1], title='Targets B', color='red')
```

For dataset B we observe that a large gap in the weather data. This is the last part of the observed weather. We also see that the target values are missing for this period. There is also a suspicion about missing target values around august 2020 and april 2021 where there are straight lines indicating something is missing.

```python
fig, axs = plt.subplots(2, 1, figsize=(20, 10), sharex=True)
feature_name = 'absolute_humidity_2m:gm3'
X_frames_train[2][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test  C', color='blue')
X_frames_test[2][['date_forecast', feature_name]].set_index('date_forecast').plot(ax=axs[0], title='Train/Test  C', color='green')
Y_frames_train[2][['time', 'pv_measurement']].set_index('time').plot(ax=axs[1], title='Targets  C', color='red')
```

For dataset C we observe the same as for dataset B, however there is also a large part of the target values missing in the start. For dataset C we would like to predict these values so that we can use them to train our model that are going to predict the target values based on X_test.
We can also notice that datasets B and C starts 6 months earlier with the data than dataset A.
